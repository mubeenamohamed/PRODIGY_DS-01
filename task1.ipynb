```python
import matplotlib.pyplot as plt

# Sample data
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan', 'Brazil', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico']
populations = [1411778724, 1393409038, 332915073, 276361783, 225199937, 213993437, 211400708, 166303498, 145912025, 130262216]

# Create bar chart
plt.figure(figsize=(12, 8))
plt.barh(countries, populations, color='skyblue')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Population Distribution by Country')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
